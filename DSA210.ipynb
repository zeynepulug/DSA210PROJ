# **Analyzing Watch Later List According to Frequency**
Because of the personal data privacy, folowing data will be analyzed without giving the content names and only will be processed in the categories of frequency of using Watch Later file in years, month, and time of the day.






```python
import pandas as pd
import matplotlib.pyplot as plt

csv_file = "/Watch later-videolar.csv"

data = pd.read_csv(csv_file)

data['Oynatma Listesi Video Oluşturma Zaman Damgası'] = pd.to_datetime(data['Oynatma Listesi Video Oluşturma Zaman Damgası'])

data['Yıl'] = data['Oynatma Listesi Video Oluşturma Zaman Damgası'].dt.year
yıllara_göre = data['Yıl'].value_counts().sort_index()

data['Ay'] = data['Oynatma Listesi Video Oluşturma Zaman Damgası'].dt.month
aylara_göre = data['Ay'].value_counts().sort_index()

plt.figure(figsize=(10, 5))
yıllara_göre.plot(kind='bar', color='skyblue')
plt.title('Video Distribution According to Years')
plt.xlabel('Year')
plt.ylabel('Number of Videos')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
plt.figure(figsize=(10, 5))
aylara_göre.plot(kind='bar', color='lightgreen')
plt.title('Video Distribution According to Months')
plt.xlabel('Month')
plt.ylabel('Number of Videos')
plt.xticks(range(12), ['January', 'February', 'March', 'April', 'May', 'June',
                       'July', 'August', 'September', 'October', 'November', 'December'], rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

# **Analysis of Subscriptions:**
At this part of the analysis thwe subscriptions have sorted. Again for the data privacy content names didn!t shown and since there is too much data they have categorized alphabetically to an easier understanding.

```python
# Gerekli kütüphaneleri yükleyin
import pandas as pd
import matplotlib.pyplot as plt

csv_file = "/content/abonelikler (1).csv"
df = pd.read_csv(csv_file)
def get_group(name):
    if pd.isna(name):
        return "Unknown"
    first_letter = name[0].upper()
    if 'A' <= first_letter <= 'F':
        return "A-F"
    elif 'G' <= first_letter <= 'K':
        return "G-K"
    elif 'L' <= first_letter <= 'P':
        return "L-P"
    elif 'Q' <= first_letter <= 'U':
        return "Q-U"
    elif 'V' <= first_letter <= 'Z':
        return "V-Z"
    else:
        return "Other"

df["Group"] = df["Kanal Başlığı"].apply(get_group)

grouped = df.groupby("Group").size().reset_index(name="Count")


grouped["Bubble Size"] = grouped["Count"]


plt.figure(figsize=(10, 6))
plt.scatter(
    grouped["Group"],
    [1] * len(grouped),
    s=grouped["Bubble Size"] * 100,
    alpha=0.6,
    c=range(len(grouped)),
    cmap="viridis",
)


for i, row in grouped.iterrows():
    plt.text(i, 1.02, f'{row["Count"]}', ha="center")


plt.title("Bubble Chart Analysis (Grouped Alphabetically)", fontsize=16)
plt.xlabel("Groups", fontsize=12)
plt.ylabel("Fixed Y Axis", fontsize=12)
plt.yticks([])


plt.show()
```
